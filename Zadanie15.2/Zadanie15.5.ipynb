```python
import pandas as pd
```

```python
prices = [
	(1, 2.12),
	(2, 2.56),
	(3, 3.10),
	(4, 3.16),
	(5, 3.58),
	(6, 5.12),
	(7, 5.16),
	(8, 5.20),
	(9, 4.12),
	(10, 4.10),
	(11, 3.65),
	(12, 4.25),
]
```

```python
df = pd.DataFrame(prices)
df = pd.DataFrame(prices, columns=["month", "price"])
df = df.set_index("month")
df
```

```python
df['priceUSD']='0'
df
```

```python
df['priceUSD']=df["price"].apply(lambda price: price / 4)
df
```

```python
import matplotlib.pyplot as plt
plt.plot(df.index, df['priceUSD'], 'r--')
plt.xlabel("months")
plt.title("Price of goods (USD)")
plt.ylim(0,2)
```
